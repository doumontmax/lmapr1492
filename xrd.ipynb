```python
from pymatgen.ext.matproj import MPRester 
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("N7ufawmcbIbaWNUI5qt") as m:
     
     # Obtenir la structure à partir du Materials Project Id
     structure = m.get_structure_by_material_id("mp-8880")
```

```python
# Créer un object XRDCalculator associé à un rayonnement X incidant CuKa de longueur d’onde λ=1.54060 Å 
XRD = XRDCalculator(1.54060)  

# Obtenir le modèle de diffraction 
pattern = XRD.get_pattern(structure)
```

```python
# Premier pic de diffraction 

# Indices hkls associés au premier pic de diffraction
hkls_pic1 = pattern.hkls[0]

# Angle du premier pic de diffraction
Angle1 = pattern.x[0]

# Intensité du premier pic de diffraction
Intensité1 = pattern.y[0]

print("Premier pic de diffraction :")
print("----------------------------")
print("Indices hkl associés : ", hkls_pic1)
print("Angle : 2 theta =", Angle1,"°")
print("Intensité :", Intensité1)
```

```python
# Deuxième pic de diffraction 

# Indices hkls associés au deuxième pic de diffraction
hkls_pic2 = pattern.hkls[1]

# Angle du deuxième pic de diffraction
Angle2 = pattern.x[1]

# Intensité du deuxième pic de diffraction
Intensité2 = pattern.y[1]

print("Deuxième pic de diffraction :")
print("----------------------------")
print("Indices hkl associés : ", hkls_pic2)
print("Angle : 2 theta =", Angle2,"°")
print("Intensité :", Intensité2)
```

```python
# Troisième pic de diffraction 

# Indices hkls associés au troisième pic de diffraction
hkls_pic3 = pattern.hkls[2]

# Angle du troisième pic de diffraction
Angle3 = pattern.x[2]

# Intensité du troisième pic de diffraction
Intensité3 = pattern.y[2]

print("Troisième pic de diffraction :")
print("----------------------------")
print("Indices hkl associés : ", hkls_pic3)
print("Angle : 2 theta =", Angle3,"°")
print("Intensité :", Intensité3)
```

```python
# Plot le diagramme de diffraction

print("Les résultats obtenus précédemment peuvent être vérifiés sur le graphe ci-dessous. En effet, les trois premiers pics de diffraction pour un rayonnement X incidant CuKa de longueur d’onde λ=1.54060 Å sont :")
XRD.get_plot(structure, two_theta_range=(20, 30), annotate_peaks=True, ax=None, with_labels=True, fontsize=10)
```
