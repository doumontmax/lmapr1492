```python
from pymatgen.ext.matproj import MPRester 
from pymatgen.io.cif import CifWriter
from pymatgen.core.operations import SymmOp

with MPRester("N7ufawmcbIbaWNUI5qt") as m:
     
     # Obtenir la structure à partir du Materials Project Id
     structure = m.get_structure_by_material_id("mp-8880")
        
     # Ecrire un fichier cif 
     w = CifWriter(structure) 
     w.write_file('mp8880.cif') 
```

```python
#1er élément de symétrie
Sym1 = SymmOp.from_xyz_string('-x+y, -x, z') 

#Matrice de rotation associée
rot1 = Sym1.rotation_matrix

#Effet de l'axe de rotation d'ordre 3 sur l'atome Al situé en (1/3, 2/3, 0)
Eff1 = Sym1.operate((1/3, 2/3, 0))
```

```python
print("1er élément de symétrie : Axe de rotation d'ordre 3 (#5 fichier cif : (-x+y, -x, z))")     
print("------------------------")

print("La matrice de rotation associée est :")
print(rot1)
print("")

print("Le vecteur (x, y, z) multiplié par la matrice de rotation donne donc bien (-x+y, -x, z) : ") 
print("")
print("Effet de l'axe de rotation d'ordre 3 sur l'atome Al situé en (1/3, 2/3, 0) : ")
print(Eff1) 

print("")
print("Illustration de l'axe de rotation d'ordre 3 sur l'atome Al situé en (1/3, 2/3, 0) :")
from IPython.display import Image
Image(filename = 'Sym1.PNG')
```

```python
#2e élément de symétrie
Sym2 = SymmOp.from_xyz_string('x, x-y, z') 

#Matrice de rotation associée
rot2 = Sym2.rotation_matrix

#Effet du plan miroir sur l'atome P situé en (1/3, 2/3, 0.37)
Eff2 = Sym2.operate((1/3, 2/3, 0.37))
```

```python
print("2e élément de symétrie : Plan miroir (#8 fichier cif : (x, x-y, z))")     
print("----------------------")

print("La matrice de rotation associée est :")
print(rot2)
print("")

print("Le vecteur (x, y, z) multiplié par la matrice de rotation donne donc bien : ") 
print("(x, x-y, z)")
print("")
print("Effet du plan miroir sur l'atome P situé en (1/3, 2/3, 0.37) : ") 
print(Eff2)


print("")
print("Illustration du plan miroir sur l'atome P situé en (1/3, 2/3, 0.37) : ") 
from IPython.display import Image
Image(filename = 'Sym2.PNG')
```

```python
#3e élément de symétrie
Sym3 = SymmOp.from_xyz_string('y, -x+y, z+1/2') 

#Matrice de rotation associée
rot3 = Sym3.rotation_matrix

#Vecteur de translation associé
vec3 = Sym3.translation_vector

#Effet de la rotation hélicoïdale sur l'atome Al situé en (2/3, 1/3, 1/2)
Eff3 = Sym3.operate((2/3, 1/3, 1/2))
```

```python
print("3e élément de symétrie : Rotation hélicoïdale (#6 fichier cif : (y, -x+y, z+1/2))")     
print("----------------------")
print("La matrice de rotation associée est :")
print(rot3)
print("")
print("Le vecteur de translation associé est :")
print(vec3)
print("")


print("Le vecteur (x, y, z) multiplié par la matrice de rotation auquel on ajoute le vecteur \n de translation associé donne donc bien : ") 
print("(y, -x+y, z+1/2)") 
print("")
print("Effet de la rotation hélicoïdale sur l'atome Al situé en (2/3, 1/3, 1/2) : ") 
print(Eff3)

print("")
print("Illustration de la rotation hélicoïdale sur l'atome Al situé en (2/3, 1/3, 1/2) : ") 
from IPython.display import Image
Image(filename = 'Sym3.PNG')
```
